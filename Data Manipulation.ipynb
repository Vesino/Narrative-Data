```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### In this notebook we'll check if the data is consistency

```python
path = './challenge_datasets/narrativewave_WTG_20201010_wtg10min_transform.csv'
```

```python
#pd.set_option('display.max_columns', None, 'display.max_rows', None)
raw_data = pd.read_csv(path)
```

```python
raw_data
```

```python
# The next part of the code help to separet each important value in the column tag and create a series using
# one of the most powerful tools in python, "list comprenhension"
asset = pd.Series([x.split('/')[2] for x in raw_data.tag])
asset
```

```python
# and so on
column = pd.Series([x.split('/')[3] for x in raw_data.tag])
column
```

```python
site = pd.Series([x.split('/')[1] for x in raw_data.tag])
site
```

```python
# A copy of Raw_data
data = raw_data
```

```python
# Replace the timestamp value on value column
data['value'] = pd.to_numeric(data['value'], errors="coerce").fillna(0).astype(int).to_frame()
```

```python
data
```

```python
# Adding three new columns this for every value on tag column
data['site'] = site
data['asset'] = asset
data['column'] = column
```

```python
# Visualization of data
data.head()
```

```python
# Creating a pivot table in order to see the value for a column by a given asset
pivot_table = pd.pivot_table(data, columns=['column'], index=['asset'], values=['value'], aggfunc=np.sum, fill_value=0)
```

```python
pivot_table
```

```python
# dummy_dframe could be our solution for a relation between two models.
data_dict = {'asset': asset, 'column': column, 'value': data['value']}
dummy_dframe = pd.DataFrame(data_dict)
dummy_dframe
```

```python
# grouped_by is a dictionary which agroup by asset the numbers of columns by asset
grouped_by = dict(list(dummy_dframe.groupby('asset')))
grouped_by
```

```python
# We could easily inspect by asset the number of columns, as follows:
# the next line returns a pandas.Dataframe
grouped_by['WTG12']
```

```python
# Since grouped_by is a group by asset we could see how many columns by asset and retrieve a number of column by asset
# The next line retreives the number of columns by the asset WTG12
# We could use the same logic in order to retreive all number of columns by a given asset

grouped_by['WTG12']['column'].value_counts()
```

```python
# unique_asset returns an array for the uniques assets values (len = 16)
unique_asset = dummy_dframe['asset'].unique()
unique_asset
```

```python
len(unique_asset)
```

```python
# asset_count returns a pd.Series with the numbers of asset in the data
# this is important because it is said by itself how many times you can see an asset in the data, or how many times repeat
asset_count = dummy_dframe['asset'].value_counts()
asset_count
```

```python
# we use the same methods like before in order to see uniques values and count them
unique_column = pd.Series(dummy_dframe['column'].unique())
unique_column
```

```python
number_of_columns = pd.Series(dummy_dframe['column'].value_counts())
number_of_columns
```
